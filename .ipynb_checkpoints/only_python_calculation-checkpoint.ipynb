```python
import numpy as np
import re
import scipy.io as spio
import math
import matplotlib.pyplot as plt
import statistics

from scipy.signal import medfilt
from scipy.interpolate import interp1d
%matplotlib notebook
```

```python
def readTextTable(file_name, head_lines=0, cmpl=False):
    e = ''
# Check comsol text file ------------------------------------------------------
    try:
        f = open(file_name, "r")
    except IOError:
        e = ("Error! Cannot open file ""%s"" !") % file_name
        Data = np.array([], float)
        return (Data,e)
    #end_try
# -----------------------------------------------------------------------------

# Read data from file ---------------------------------------------------------
    lines_all = f.readlines()                                                   # read all string lines in the file and save to the list 'lines_all'
    f.close()

    str_list = re.split('\s+', lines_all[head_lines])                           # split line into 'words'
    str_list = list(filter(None, str_list))                                     # remove '' from the list

    cols = len(str_list)
    rows = len(lines_all)-head_lines

    Data = np.zeros((rows, cols), complex)                                   # allocate memory for data

    line_idx = head_lines
    i = 0
    for line in lines_all[head_lines:]:
        str_list = re.split('\s+', line.replace('i', 'j'))                      # split line into 'words'
        str_list = list(filter(None, str_list))                                 # remove '' from the list
        try:                                                                    # check whether all words in the line are numeric values
            data_list = [complex(s) for s in str_list]
        except (ValueError, IndexError):
            e += "Incorrect data in line %d !" % (line_idx)
            if not cmpl: Data = np.real(Data)
            return (Data,e)
        #end_try

        if len(data_list) == cols:
            Data[i,:] = np.array(data_list, complex)
        else:
            e += "Incorrect data in line %d !" % (line_idx)
            if not cmpl: Data = np.real(Data)
            return (Data,e)
        #end_if

        i += 1
        line_idx += 1
    #end_for

    if not cmpl: Data = np.real(Data)
    return (Data,e)


def dBToPow(data):
    return 10**(data/10)
```

## Сырой формат данных

```python
fname_lfp = r"C:\TGT\1.02.2022\Averaging_SNL10a_1461_24.11.2021-LFP.txt"
fname_hfp = r"C:\TGT\1.02.2022\Averaging_SNL10a_1461_24.11.2021-HFP.txt"
```

```python
Q = [0, 0, 15, 5, 10.3, 15.2, 17.2, 10.7, 22.7, 30.5, 40.6, 50.2, 61.6, 70, 79.7, 92, 103.3, 123.8, 141.4, 160 ]
Q = np.array(Q, float)

# Read signal + background + resonant noise data (LFP)
raw_lfp,_ = readTextTable(fname_lfp)

freq_lfp = raw_lfp[0,:]
raw_lfp = raw_lfp[1:,:].squeeze()


# Read signal + background + resonant noise data (HFP)
raw_hfp,_ = readTextTable(fname_hfp)

freq_hfp = raw_hfp[0,:]
raw_hfp = raw_hfp[1:,:].squeeze()

qc_idx = [True, True, False, False, False, False, True, True, True, True, True, True, True, True, True, False, True, True, True, True]
Q = Q[qc_idx]
raw_hfp = raw_hfp[qc_idx]
raw_lfp = raw_lfp[qc_idx]


rate_idx = np.argsort(Q)
raw_hfp = raw_hfp[rate_idx]
raw_lfp = raw_lfp[rate_idx]
Q = Q[rate_idx]

raw_lfp = dBToPow(raw_lfp)
raw_lfp_bkg = raw_lfp[1,:] 
raw_lfp = raw_lfp[1:,:]
raw_hfp = dBToPow(raw_hfp)   
raw_hfp_bkg = raw_hfp[1,:] 
raw_hfp = raw_hfp[1:,:]
Q = Q[1:]


stations = len(Q)
lfp = raw_lfp - np.tile(raw_lfp_bkg, [stations,1])
hfp = raw_hfp - np.tile(raw_hfp_bkg, [stations,1])

idx_4kHz = freq_hfp >= 4.0 
hfp = hfp[:,idx_4kHz]
freq_hfp_res = freq_hfp[idx_4kHz]

freq_lfp_new = freq_hfp[freq_hfp < 4.0]
interp = interp1d(freq_lfp, lfp)
lfp_new = interp(freq_lfp_new)

hfp_new = np.hstack((lfp_new, hfp))

fig, ax = plt.subplots()
ax.set_yscale('log')
ax.grid()

for raw in raw_lfp:
    ax.plot(freq_lfp, raw)
```

### Уберем точки, которые после вычитания бэкграунда остались с амплитудой меньше или равно нуля

```python
for i in range(1, len(lfp)):
    freq_lfp_after_del = freq_lfp[lfp[i] > 0]
    interp = interp1d(freq_lfp_after_del, lfp[i][lfp[i] > 0], fill_value="extrapolate")
    lfp[i] = interp(freq_lfp)

for i in range(1, len(hfp)):
    freq_hfp_after_del = freq_hfp_res[hfp[i] > 0]
    interp = interp1d(freq_hfp_after_del, hfp[i][hfp[i] > 0], fill_value="extrapolate")
    hfp[i] = interp(freq_hfp_res)
    
plt.plot(freq_lfp, lfp[1])
plt.yscale('log')
plt.grid()
```

## Определение модельных матриц
### Для НЧ спектра Q = 10,7 m3/min, ВЧ Q = 30,5 m3/min

```python
win = 100
filt = np.ones(win)/win

res_lfp = np.convolve(medfilt(lfp[1], 49), filt, mode='same')[50:-50]
res_hfp = np.convolve(medfilt(hfp[4], 49), filt, mode='same')

model_lfp = np.array([res_lfp, np.array([1 for i in range(len(res_lfp))])])
model_lfp = model_lfp.T

model_hfp = np.array([res_hfp, np.array([1 for i in range(len(res_hfp))])])
model_hfp = model_hfp.T
```

## Предполагаем линейную связь между модельной матрицей и экспериментальными данными
### Решая матричные уравнения ищем набор коэффициентов А, В

```python
from scipy.linalg import pinv

s = pinv(model_lfp)
A = []
B = []

for lfp_one in lfp:
    A0 = np.matmul(s, lfp_one[50:-50])
    A.append(A0[0])
    B.append(A0[1])

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(Q, A, label='Q(A)')
ax1.set_yscale('log')
ax1.set_xlabel('Q, m3/min')
ax1.set_ylabel('A')
ax1.grid()
ax1.legend()

ax2.scatter(Q, B, label='Q(B)')
ax2.set_yscale('log')
ax2.set_xlabel('Q, m3/min')
ax2.set_ylabel('B')
ax2.grid()
ax2.legend()
```

## Описание моделью экспериментальных данных в НЧ области 

```python
# Q[5] = 40.6
i = 4

A_i = np.array([np.array([A[i]]), np.array([B[i]])])
exp_lfp = np.matmul(model_lfp, A_i)

fig, ax = plt.subplots()

ax.plot(freq_lfp[50:-50], lfp[i][50:-50], label='Эксперимент')
ax.plot(freq_lfp[50:-50], exp_lfp, label='Модель')

ax.grid()
ax.set_yscale('log')
ax.legend()
```
